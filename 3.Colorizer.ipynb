# Colorize De Boer images from drive

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
root_dir = '/content/drive/MyDrive/dating-images/'
input_folder = root_dir + '/data/images/'
output_folder = root_dir +  '/data/images-colorized/'

import glob

image_paths = glob.glob(input_folder + '/**/*.jpg')
print(len(image_paths))
```

```python
from PIL import Image
import PIL
import os

render_factor = 35

for path in image_paths:
  scene = path.split('/')[-2]
  os.makedirs(output_folder + '/' + scene, exist_ok=True)

  filename = path.split('/')[-1]

  image = colorizer.get_transformed_image(path=path, render_factor=render_factor, post_process=True, watermarked=False)
  image = image.save(output_folder + '/' + scene + '/' + filename)
```
